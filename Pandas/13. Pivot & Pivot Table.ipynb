# Pivot

## Dataset

```python
import pandas as pd

data = [
    ['1/1/2017', 'new york', 32, 6, 'Rain'],
    ['1/2/2017', 'new york', 36, 7, 'Sunny'],
    ['1/3/2017', 'new york', 28, 12, 'Snow'],
    ['1/4/2017', 'new york', 33, 7, 'Sunny'],
    ['1/1/2017', 'mumbai', 90, 5, 'Sunny'],
    ['1/2/2017', 'mumbai', 85, 12, 'Fog'],
    ['1/3/2017', 'mumbai', 87, 15, 'Fog'],
    ['1/4/2017', 'mumbai', 92, 5, 'Rain'],
    ['1/1/2017', 'paris', 45, 20, 'Sunny'],
    ['1/2/2017', 'paris', 50, 13, 'Cloudy'],
    ['1/3/2017', 'paris', 54, 8, 'Cloudy'],
    ['1/4/2017', 'paris', 42, 10, 'Cloudy'],
]

df = pd.DataFrame(data, columns = ['day', 'city', 'temperature', 'windspeed', 'event'])
df
```

## Pivot entire DataFrame

```python
df.pivot(index = 'day', columns = 'city')
```

## Include specific columns in Pivoted DataFrame

```python
df.pivot(index = 'day', columns = 'city', values = ['temperature', 'windspeed'])
```

# Pivot Table

## Dataset

```python
data = [
    ["5/1/2017", "new york", 65, 56],
    ["5/1/2017", "new york", 61, 54],
    ["5/2/2017", "new york", 70, 60],
    ["5/2/2017", "new york", 72, 62],
    ["5/1/2017", "mumbai", 75, 80],
    ["5/1/2017", "mumbai", 78, 83],
    ["5/2/2017", "mumbai", 82, 85],
    ["5/2/2017", "mumbai", 80, 26],
]

df = pd.DataFrame(data, columns=["date", "city", "temperature", "humidity"])
df
```

## Calculates datewise Avg by default

```python
df.pivot_table(index = 'city', columns = 'date')
```

## `aggfunc` argument:
- To calculate `sum`

```python
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum')
```

- To calculate `count`

```python
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'count')
```

- To calculate `max`

```python
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'max')
```

## `Grouper():`
- Used to do groupBy & pivot table simultaneously

## Dataset

```python
data = [
    ["5/1/2017", "new york", 65, 56],
    ["5/2/2017", "new york", 61, 54],
    ["5/3/2017", "new york", 70, 60],
    ["12/1/2017", "new york", 30, 50],
    ["12/2/2017", "new york", 28, 52],
    ["12/3/2017", "new york", 25, 51],
]

df = pd.DataFrame(data, columns=["date", "city", "temperature", "humidity"])
df
```

## Calculate month-wise avg values & pivot table

```python
df['date'] = pd.to_datetime(df['date'])
df.pivot_table(index = pd.Grouper(freq = 'M', key = 'date'), columns = 'city')
```

## Dataset

```python
import pandas as pd

df = pd.read_csv("aapl_no_dates.csv")
df.head(15)
```

## Insert Date column using `date_range`
- Input the start & end date
- Input the frequency. (B -> Business Days)
- As the stocks aren't traded on weekends we used frequency as `B`

```python
rng = pd.date_range(start = "6/1/2016", end = "6/30/2016", freq = 'B')
df.set_index(rng, inplace = True)
df.head(15)
```

## `asfreq()`
- To get the weekend data we can copy next or previous data

### ffill/pad:

```python
# To copy Friday's data in weekend
df.asfreq('D', 'pad').head(15)
```

### bfill/backfill

```python
# To copy Monday's data in weekend
df.asfreq('D', 'bfill').head(15)
```

# Dataset

```python
import pandas as pd

data = [
    ['1/1/2017', '32 F', '6 mph', 'Rain'],
    ['1/2/2017', -99999, '7 mph', 'Sunny'],
    ['1/3/2017', '28 c', -99999, 'Snow'],
    ['1/4/2017', -99999, 7, 0],
    ['1/5/2017', 32, -99999, 'Rain'],
    ['1/6/2017', 31, 2, 'Sunny'],
    ['1/6/2017', 0, 5, 0],
]

df = pd.DataFrame(data, columns = ['day', 'temperature', 'windspeed', 'event'])
df
```

- Replace -99999 with 28

```python
df.replace(-99999, 28)
```

- Column specific value replace

```python
df = df.replace({'temperature': -99999, 'windspeed': -99999, 'event': 0}, value = {'temperature': 28, 'windspeed': 5, 'event': 'Sunny'})
df
```

- Chop off `alphabets` from temperature & windspeed column & only `keep digits` using `REGEX`

```python
df.replace({'temperature' : '[A-Za-z ]', 'windspeed' : '[A-Za-z ]'}, '', regex = True)
```
